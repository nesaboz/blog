# Paperspace.com

[Paperspace](https://www.paperspace.com/) offers very affordable remote machines with GPU, and allows for 150Gb permanent storage. I find it better easy to set up platform then any other.
It also has many pre-installed deployments PyTorch, FastAI, Stable Diffusion, and more. It does require monthly subscription that is at the time of writting $8 per month.

# Gradient ML platform

For most of my needs, I use free machines on Gradient:

don't forget to set auto-shutdown depending on you need, 6 hours is the max:



Once the machine is running, the small icon allows to get the URL of the jupyter kernel on a remote server.





I then copy the URL in my local VSCode, navigate to "Select Kernel"->"Select another Kernel"->"Existing Jupyter Kernel"-> Enter the URL of the running Jupyter server" -> "Python3 (ipykernel)":




That's it. The kernel is now set in the upper right corner of the VSCode:



[Link](https://docs.paperspace.com/gradient/notebooks/notebooks-remote-kernel/) on paperspace docs one can follow to trouble shoot the setup.

This is great for majority of my needs, one can always restart the machine after 6 hours.

All files in the /notebook folder are persistent, outside that no.

# Core machines



I haven't seen an option to SSH into these Gradient machines but this option exist for Core servers. The ones with GPU cost $0.5 per hour, but for sure one can SSH into those.
