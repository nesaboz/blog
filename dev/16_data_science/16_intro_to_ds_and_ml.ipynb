Inferential statistics: sample, population, hypothesis testing, confidence interval, regression analysis

Joint probability: random variable, independent events, mutually exclusive events, non-mutually exclusive, conditional probability,  Bayes theorem (posterior probability, prior probability, likelihood, evidence)

Probability distribution: discrete/continuous random variable is described by probability mass/density function 
Bernoulli, binomial, normal, standard normal, Poisson 
Skewness and kurtosis (3rd and 4th standardized moment)

"Regression models in Machine Learning are affected by the presence of outliers, which can be indicated from a skewed distribution. It becomes necessary at times to remove the Skewness. Taking the log transform of the respective column of a dataset can also help reduce the skewness. There is a famous Box-Cox Method, which takes in a parameter to transform the column of a dataset so it is more Gaussian-like."

Sampling: random, stratified

Key concepts: Population vs sample, statistical inference (estimation and statistical hypothesis testing), parameter vs statistics (sample error), central limit theorem (if population distribution is $X \sim \mathcal{N}(\mu,\,\sigma^{2})$ then distribution of sample means is $Y \sim \mathcal{N}(\mu,\,\frac{\sigma^{2}}{n})$).

Statistical Hypothesis testing (null hypothesis, alternative hypothesis, type 1 error = false positive = significance level = $\alpha$, type 2 error = false negative = $\beta$, degrees of freedom

p-value approach, critical value approach, t-test (`scipy.stats.ttest_ind`)

Recomender systems: content-based, collaborative filtering (user-based, item-based).

Hybrid approaches.

### Regression

independet vs dependent features, rows are samples
datasets: train, validation, test
numerical vs categorical columns, ordinal (has order) vs nominal (no order).
underfitting (high-bias, training dataset performance is low), overfitting (high-variance, training dataset performance is high, validation dataset performance is low).
