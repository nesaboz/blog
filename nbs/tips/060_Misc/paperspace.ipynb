# Paperspace.com

[Paperspace](https://www.paperspace.com/) offers very affordable remote machines with GPU, and allows for 150Gb permanent storage (that is mounted as `/notebook` folder). I find it easier to set up then any other platform.
It has many pre-installed deployments like PyTorch, FastAI, Stable Diffusion, and more. It does require monthly subscription that is at the time of writting $8 per month.

That's pretty much it; the kernel is now set in the upper right corner of the VSCode:

This is great for majority of my needs, for longer durations one can always restart the machine.

Note again that only files in the `/notebook` folder are persistent.
